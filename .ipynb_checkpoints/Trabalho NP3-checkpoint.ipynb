```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from sklearn import datasets

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, f1_score

from sklearn.metrics import roc_curve, auc
from sklearn import datasets
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

from itertools import cycle
from scipy import interp
```

```python
file = pd.read_csv("mushrooms.csv")
```

```python
file
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for i in file.columns:
    file[i] = label_encoder.fit_transform(file[i])
file.head()
```

```python
X = pd.DataFrame(file, columns=['cap-surface','cap-color']) 
y = pd.DataFrame(file, columns=['class'])
```

```python
X
```

```python
y
```

<H1> Regressão Logística

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print("TREINAMENTO:")
print("X:", X_train.shape, "y:", y_train.shape)
print("TESTE:")
print("X:", X_test.shape, "y:", y_test.shape)
```

```python
model = Pipeline([('nor', MinMaxScaler()), ('LR', OneVsRestClassifier(LogisticRegression(multi_class='ovr')))])
params = {}
```

```python
X_train
```

```python
gs = GridSearchCV(model, params, cv=5, refit=True)
gs.fit(X_train, y_train)
```

```python
gs.cv_results_
```

```python
y_pred = gs.predict(X_test)
```

```python
y_test
```

```python
type(y_test)
```

```python
print("Accuracy")
print(accuracy_score(y_test, y_pred))
#print("confusion matrix")
#confusion_matrix(y_test["class"].argmax(), y_pred.argmax())
```
